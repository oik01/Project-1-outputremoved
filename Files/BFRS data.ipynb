```python
import pandas as pd
```

```python
BFRS_2016= "/Users/omarkreidieh/Documents/GitHub/Flu-Project/Resources/BRFSS/BRFSS2016.xpt"
BFRS_2016_df= pd.read_sas(BFRS_2016,format='xport', index=None, encoding='ISO-8859-1', chunksize=None, iterator=False)
BFRS_2016_df.head()
```

```python
# Rename and select the columns that I need to use: 
BFRS_2016_df_renamed= BFRS_2016_df.rename(columns={"_STATE":"State", "IYEAR": "Year", "AGE": "Age", "_AGEG5YR": "Age-group", "_RACE": "Race", "CVDINFR4": "Myocardial Infarction", "CVDCRHD4": "Coronary Artery Disease", "CVDSTRK3": "Stroke", "CHCCOPD1": "COPD", "CHCKIDNY": "Kidney Disease", "DIABETE3": "Diabetes", "SEX": "Gender", "FLUSHOT6": "Flu Vaccine", "FLSHTMY2":"Flu Vaccine Date", "SMOKE100": "Smoking", "_LLCPWT": "Weighting Variable"})
BFRS_2016_df_selected=BFRS_2016_df_renamed[["State", "Year", "Gender", "Age-group", "Race", "Smoking", "Myocardial Infarction", "Coronary Artery Disease","Stroke", "COPD", "Kidney Disease", "Diabetes", "Flu Vaccine", "Flu Vaccine Date", "Weighting Variable"]]
merged_df = pd.DataFrame(BFRS_2016_df_selected)
merged_df.head()

print(len(merged_df["Gender"]))

merged_df.Year.unique()
```

```python
# merge other files to the 2016 dataframe
other_files = ["/Users/omarkreidieh/Documents/GitHub/Flu-Project/Resources/BRFSS/BRFSS2015.xpt", "/Users/omarkreidieh/Documents/GitHub/Flu-Project/Resources/BRFSS/LLCP2014.xpt", "/Users/omarkreidieh/Documents/GitHub/Flu-Project/Resources/BRFSS/BRFSS2013.xpt","/Users/omarkreidieh/Documents/GitHub/Flu-Project/Resources/BRFSS/LLCP2012.xpt"]

for file in other_files:
    new_df = pd.read_sas(file)
    new_renamed_df = new_df.rename(columns={"_STATE":"State", "IYEAR": "Year", "AGE": "Age", "_AGEG5YR": "Age-group", "CVDINFR4": "Myocardial Infarction", "CVDCRHD4": "Coronary Artery Disease", "CVDSTRK3": "Stroke", "CHCCOPD1": "COPD", "CHCKIDNY": "Kidney Disease", "DIABETE3": "Diabetes", "SEX": "Gender", "FLUSHOT6": "Flu Vaccine", "FLSHTMY2":"Flu Vaccine Date", "SMOKE100": "Smoking", "_LLCPWT": "Weighting Variable", 'FLUSHOT5': 'Flu Vaccine'})
    new_selected=new_renamed_df[["State", "Year", "Gender", "Age-group", "Smoking", "Myocardial Infarction", "Coronary Artery Disease","Stroke", "COPD", "Kidney Disease", "Diabetes", "Flu Vaccine", "Flu Vaccine Date", "Weighting Variable"]]
    merged_df = pd.concat([new_selected, merged_df])
    merged_df.reset_index()
    print(f'finished with {file}')

print(len(merged_df["Gender"]))
    
```

```python
# export to a csv file 

merged_df.to_csv("/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/TotalBRFSS.csv", index = False, header = True)
```
