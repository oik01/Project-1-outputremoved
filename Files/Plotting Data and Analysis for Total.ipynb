```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
icd_9_total = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/ICD_9_Totalincidence.csv')
```

```python
icd_9_2015_total = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/2015_Totalincidence.csv')
icd_10_2015_total = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/2015_Q4Totalincidence.csv')
```

```python
total_2015 = pd.merge(icd_9_2015_total, icd_10_2015_total, how = 'outer' )
```

```python
total_2015 = total_2015.groupby(['YEAR','AMONTH']).sum()
```

```python
total_2015 = total_2015.reset_index()
total_2015 = total_2015[['YEAR', 'AMONTH', 'Influenza Incidence','MI Incidence' ]]
total_2015
```

```python
df_2016 = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/2016Totalincidence.csv')
                      
```

```python
df_2016
```

```python
NIS_data = pd.concat([icd_9_total, total_2015, df_2016 ])
```

```python
NIS_data = NIS_data[['YEAR', 'AMONTH', 'Influenza Incidence', 'MI Incidence']]
NIS_data= NIS_data.reset_index(drop = True)
```

```python
NIS_data
```

```python
Months =[]
index_list = list(NIS_data.index.values)
for index in index_list:
    year = NIS_data['YEAR'][index]
    Month_no_year = NIS_data['AMONTH'][index]
    month= f'{year}{Month_no_year}'
    Months.append(month)

    
```

```python
print(Months)
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Time')
ax1.set_ylabel('Influenza Monthly Incidence', color = 'blue')
ax1.plot(Months,NIS_data['Influenza Incidence'], color = 'blue', label = 'Influenza' )
ax1.tick_params(axis = 'y', labelcolor = 'blue')
ax2= ax1.twinx()
ax2.set_ylabel('Myocardial Infarction Incidence', color = 'red')
ax2.plot(Months,NIS_data['MI Incidence'], color = 'red', label = 'Myocardial Infarction (MI)')
ax2.tick_params(axis='y', labelcolor= 'red')
ax1.xaxis.set_major_locator(plt.MaxNLocator(5))

fig.legend(bbox_to_anchor=(1.2, 1), loc=2)
fig.suptitle('Influenza and Myocardial Infarction Monthly Events over Time', fontsize=16)

```

```python
vaccine_effectiveness = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/cdc_effectiveness.csv')
vaccine_effectiveness
```

```python
# Assign Vaccine Effectiveness to each month using the np.select method with multiple conditionals
later_year= [7,8,9,10, 11, 12]
early_year= [1,2,3,4,5,6]
Conditions = [ (NIS_data['YEAR'] == 2012) & (NIS_data['AMONTH'].isin(early_year)), (NIS_data['YEAR'] == 2012) & (NIS_data['AMONTH'].isin(later_year)), (NIS_data['YEAR'] == 2013) & (NIS_data['AMONTH'].isin(early_year)), (NIS_data['YEAR'] == 2013) & (NIS_data['AMONTH'].isin(later_year)), (NIS_data['YEAR'] == 2014) & (NIS_data['AMONTH'].isin(early_year)), (NIS_data['YEAR'] == 2014) & (NIS_data['AMONTH'].isin(later_year)), (NIS_data['YEAR'] == 2015) & (NIS_data['AMONTH'].isin(early_year)), (NIS_data['YEAR'] == 2015) & (NIS_data['AMONTH'].isin(later_year)), (NIS_data['YEAR'] == 2016) & (NIS_data['AMONTH'].isin(early_year)), (NIS_data['YEAR'] == 2016) & (NIS_data['AMONTH'].isin(later_year))]
choices = [47, 49, 49, 52, 52, 19, 19, 48, 48, 40 ]
NIS_data['Vaccine_Effectiveness'] = np.select(Conditions, choices, default='not yet determined')
```

```python
NIS_data.dtypes
NIS_data['Vaccine_Effectiveness'] = NIS_data['Vaccine_Effectiveness'].astype(int)
NIS_data.dtypes
```

```python
fig2, ax1 = plt.subplots()
ax1.set_xlabel('Time- monthly')
ax1.set_ylabel('Influenza Monthly Incidence', color = 'blue')
ax1.plot(Months,NIS_data['Influenza Incidence'], color = 'blue', label = 'Influenza' )
ax1.tick_params(axis = 'y', labelcolor = 'blue')
ax2= ax1.twinx()
ax2.set_ylabel('cdc_vaccine_effectivenes', color = 'green')
ax2.plot(NIS_data['Vaccine_Effectiveness'], color = 'green', label = 'Vaccine Effectiveness')
ax2.tick_params(axis='y', labelcolor= 'green')
fig.legend(bbox_to_anchor=(1.2, 1), loc=2)
```
