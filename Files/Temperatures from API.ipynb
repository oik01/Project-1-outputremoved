```python
index_list = list(new_df.index.values)

Division_2 = ['New York', 'Pennsylvania', 'New Jersey']
Division_3 = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio']
Division_4 = ['Missouri', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa']
Division_5 = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida']
Division_6 = ['Kentucky', 'Tennessee', 'Mississippi', 'Alabama']
Division_7 = ['Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
Division_8 = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico']
Division_9 = ['Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']

new_df = new_df.reindex(new_df.columns.tolist() + ['HOSP_DIV', 'State Population'], axis=1)

for index in index_list:
    series = new_df['State'].isin(['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut'])
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 1
    if new_df['State'].isin(Division_2)[index] == True:
        new_df['HOSP_DIV'][index] = 2
    series = new_df['State'].isin(Division_3)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 3
    series = new_df['State'].isin(Division_4)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 4
    series = new_df['State'].isin(Division_5)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 5
    series = new_df['State'].isin(Division_6)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 6
    series = new_df['State'].isin(Division_7)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 7
    series = new_df['State'].isin(Division_8)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 8
    series = new_df['State'].isin(Division_9)
    if series[index] == True:
        new_df['HOSP_DIV'][index] = 9

```
