```python
import pandas as pd
```

```python
BRFSS_df= pd.read_csv("/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/TotalBRFSS.csv")
BRFSS_df.head()
```

```python
BRFSS_df['Year']=BRFSS_df['Year'].replace({"b'2013'": "2013", "b'2014'": "2014","b'2015'":"2015", "b'2016'": "2016" })
BRFSS_df['Year']=pd.to_numeric(BRFSS_df["Year"])
BRFSS_groupby = BRFSS_df.groupby(["Year", "Flu Vaccine"],as_index = False)
new_df= pd.DataFrame(BRFSS_groupby.count())
new_df
```

```python
new_df=new_df[["Year","Flu Vaccine","Weighting Variable"]]
new_df.set_index('Year')
Yearly=new_df["Weighting Variable"].sum(level="Year")
yearly_sum= pd.DataFrame(Yearly)
yearly_sum=yearly_sum.rename(columns= {"Weighting Variable" : "Yearly Total"})
new_df.reset_index(level=0,inplace = True, col_level = 0)
withyearlytotal= pd.merge(new_df,yearly_sum, on = "Year")
```
