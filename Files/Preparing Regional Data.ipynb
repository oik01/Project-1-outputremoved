```python
BRFSS_df = pd.read_csv('/Users/omarkreidieh/Documents/GitHub/Flu-Project/Output Files/TotalBRFSS.csv')
len(BRFSS_df['State'])
```

```python
BRFSS_df = BRFSS_df.dropna(subset=['State'])
len(BRFSS_df['State'])
```

```python
BRFSS_df = BRFSS_df.reindex(BRFSS_df.columns.tolist() + ['HOSP_DIV'], axis=1)
Division_1 = [23, 33, 50, 25, 44, 9]
Division_2 = [36, 42, 34]
Division_3 = [55, 26, 17, 18, 39]
Division_4 = [29, 38, 46, 31, 20, 27, 19]
Division_5 = [10, 24, 11, 51, 54, 37, 45, 13, 12]
Division_6 = [21, 47, 28, 1]
Division_7 = [40, 48, 5, 22]
Division_8 = [16, 30, 56, 32, 49, 8, 4, 35]
Division_9 = [2, 53, 41, 6, 15]
```

```python


#series_1 = BRFSS_df['State'].isin(Division_1)
#print('calculated series_1')
#series_2 = BRFSS_df['State'].isin(Division_2)
#print('calculated series_2')
#series_3 = BRFSS_df['State'].isin(Division_3)
#print('calculated series_3')
#series_4 = BRFSS_df['State'].isin(Division_4)
#print('calculated series_4')
#series_5 = BRFSS_df['State'].isin(Division_5)
#print('calculated series_5')
#series_6 = BRFSS_df['State'].isin(Division_6)
#print('calculated series_6')
#series_7 = BRFSS_df['State'].isin(Division_7)
#print('calculated series_7')
#series_8 = BRFSS_df['State'].isin(Division_8)
#print('calculated series_8')
#series_9 = BRFSS_df['State'].isin(Division_9)
#print('calculated series_9')

for index in tqdm(range(0,6553562)):
    try:
        if BRFSS_df['State'].isin(Division_1)[index] == True:
            BRFSS_df['HOSP_DIV'][index] = 1
        elif BRFSS_df['State'].isin(Division_2)[index] == True:
            BRFSS_df['HOSP_DIV'][index] = 2
        elif BRFSS_df['State'].isin(Division_3)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 3
        elif BRFSS_df['State'].isin(Division_4)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 4
        elif BRFSS_df['State'].isin(Division_5)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 5
        elif BRFSS_df['State'].isin(Division_6)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 6
        elif BRFSS_df['State'].isin(Division_7)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 7
        elif BRFSS_df['State'].isin(Division_8)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 8
        elif BRFSS_df['State'].isin(Division_9)[index]  == True:
            BRFSS_df['HOSP_DIV'][index] = 9
        else: BRFSS_df['State'][index] = ''
    except:
        print(index)
```

```python
d = {1 :[23, 33, 50, 25, 44, 9], 2 : [36, 42, 34], 3: [55, 26, 17, 18, 39], 4: [29, 38, 46, 31, 20, 27, 19], 5 : [10, 24, 11, 51, 54, 37, 45, 13, 12], 6 : [21, 47, 28, 1], 7 : [40, 48, 5, 22], 8 : [16, 30, 56, 32, 49, 8, 4, 35], 9: [2, 53, 41, 6, 15]}

d_rev = {w: k for k, v in d.items() for w in v}

BRFSS_df['HOSP_DIV'] = BRFSS_df['State'].map(d_rev)
```

```python
# I will need to do age later
Interest= ['Flu Vaccine', 'COPD', 'Coronary Artery Disease', 'Diabetes', 'Gender', 'Kidney Disease', 'Myocardial Infarction', 'Smoking', 'Stroke'   ]

def wprop(x, col):
    return (x['Weighting Variable']*(x[col]==1)).sum()/x['Weighting Variable'].sum()
grouped = BRFSS_df.groupby(['Year','HOSP_DIV'])    
new_df = pd.concat([pd.DataFrame(grouped.apply(wprop,col)) for col in Interest], axis = 1 )

#grouped = df.groupby('race')
#pd.concat([grouped.apply(wavg,col) for col in df.columns if col.startswith('Question')],axis=1)\
 #   .rename(columns = {num:f'Question_{num+1}' for num in range(3)})
#wp = lambda x: (BRFSS_df.loc[(x == 1).index,'Weighting Variable'].sum()) / (BRFSS_df.loc[x,'Weighting Variable'].sum())
#dictionary_functions = {'COPD': wp, 'Flu Vaccine': wp}
#BRFSS_group = pd.DataFrame(BRFSS_df.groupby(['Year','HOSP_DIV']).agg(dictionary_functions))
```

```python
new_df = new_df.set_axis(Interest, axis=1, inplace=False)
len(new_df['Flu Vaccine'])
```

```python
df_repeated_with_index = pd.concat([new_df]*12)
```

```python
df_repeated_with_index = df_repeated_with_index.sort_index()
df_repeated_with_index['Month'] = np.arange(len(df_repeated_with_index))
```

```python
df_repeated_with_index= df_repeated_with_index.drop(labels= '2017', axis=0, level='Year')
```
