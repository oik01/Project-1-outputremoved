```python
for file in NIS_Files:
    NISchunk= pd.read_sas(file,format='sas7bdat', index=None, encoding='ISO-8859-1', chunksize= 100000)
    chunk_list = []  # append each chunk df here 
    print(f'Created Chunk List for: {file}' )

# Each chunk is in df format
    for chunk in NISchunk:  
        chunk_list.append(chunk[['AMONTH', 'DISCWT', 'DXCCS1', 'DXCCS2', 'DXCCS3', 'DXCCS4', 'DXCCS5', 'DXCCS6', 'DXCCS7', 'DXCCS8', 'DXCCS9', 'DXCCS10', 'DXCCS11', 'DXCCS12', 'DXCCS13', 'DXCCS14', 'DXCCS15', 'DXCCS16', 'DXCCS17', 'DXCCS18', 'DXCCS19', 'DXCCS20', 'DXCCS21', 'DXCCS22', 'DXCCS23', 'DXCCS24', 'DXCCS25', 'HOSP_DIVISION', 'YEAR']])
    print(f'Appended Chunk List for: {file}' )
# concat the list into dataframe 
    df_concat = pd.concat(chunk_list)
    print(f'created the big dataframe for: {file}' )
    #Create a column specifying if the discharge had a diagnosis of Influenza
    Influenza= ['123']
    df_concat['Influenza'] = df_concat['DXCCS1'].isin(Influenza)| df_concat['DXCCS2'].isin(Influenza)| df_concat['DXCCS3'].isin(Influenza)| df_concat['DXCCS4'].isin(Influenza)| df_concat['DXCCS5'].isin(Influenza)| df_concat['DXCCS6'].isin(Influenza)| df_concat['DXCCS7'].isin(Influenza)| df_concat['DXCCS8'].isin(Influenza)| df_concat['DXCCS9'].isin(Influenza)| df_concat['DXCCS10'].isin(Influenza)| df_concat['DXCCS11'].isin(Influenza)| df_concat['DXCCS12'].isin(Influenza)| df_concat['DXCCS13'].isin(Influenza)| df_concat['DXCCS14'].isin(Influenza)| df_concat['DXCCS15'].isin(Influenza)| df_concat['DXCCS16'].isin(Influenza)| df_concat['DXCCS17'].isin(Influenza)| df_concat['DXCCS18'].isin(Influenza)| df_concat['DXCCS19'].isin(Influenza)| df_concat['DXCCS20'].isin(Influenza)| df_concat['DXCCS21'].isin(Influenza)| df_concat['DXCCS22'].isin(Influenza)| df_concat['DXCCS23'].isin(Influenza)| df_concat['DXCCS24'].isin(Influenza)| df_concat['DXCCS25'].isin(Influenza)    
    print(f'created a dataframe with influenza admissions for : {file}' )

    # Create a column specifying if the discharge had a diagnosis of myocardial infarction
    MI = ['100']
    df_concat['MI'] = df_concat['DXCCS1'].isin(MI)| df_concat['DXCCS2'].isin(MI)| df_concat['DXCCS3'].isin(MI)| df_concat['DXCCS4'].isin(MI)| df_concat['DXCCS5'].isin(MI)| df_concat['DXCCS6'].isin(MI)| df_concat['DXCCS7'].isin(MI)| df_concat['DXCCS8'].isin(MI)| df_concat['DXCCS9'].isin(MI)| df_concat['DXCCS10'].isin(MI)| df_concat['DXCCS11'].isin(MI)| df_concat['DXCCS12'].isin(MI)| df_concat['DXCCS13'].isin(MI)| df_concat['DXCCS14'].isin(MI)| df_concat['DXCCS15'].isin(MI)| df_concat['DXCCS16'].isin(MI)| df_concat['DXCCS17'].isin(MI)| df_concat['DXCCS18'].isin(MI)| df_concat['DXCCS19'].isin(MI)| df_concat['DXCCS20'].isin(MI)| df_concat['DXCCS21'].isin(MI)| df_concat['DXCCS22'].isin(MI)| df_concat['DXCCS23'].isin(MI)| df_concat['DXCCS24'].isin(MI)| df_concat['DXCCS25'].isin(MI)
    print(f'created a dataframe with MI admissions for : {file}' )
    sum_df = pd.DataFrame(df_concat.loc[df_concat['Influenza']== True].groupby('AMONTH').sum())
    # Create a new data frame object grouped by region and month and only including influenza discharges
    Influenza_Regional = pd.DataFrame(df_concat.loc[df_concat['Influenza']== True].groupby(['YEAR','HOSP_DIVISION','AMONTH']).sum())
    Influenza_Regional = Influenza_Regional[['DISCWT']]
    Influenza_Regional = Influenza_Regional.rename(columns = {'DISCWT': 'Influenza Incidence'})
    print(f'created a dataframe with only regional Influenza admissions incidence : {file}' )

    # Same process without region for a total incidence 
    Influenza_Total = pd.DataFrame(df_concat.loc[df_concat['Influenza']== True].groupby(['YEAR','AMONTH']).sum())
    Influenza_Total = Influenza_Total[['DISCWT']]
    Influenza_Total = Influenza_Total.rename(columns = {'DISCWT': 'Influenza Incidence'})
    Influenza_Total.head()
    print(f'created a dataframe with only total Influenza admissions incidence : {file}' )

    # MI Regional
    MI_Regional = pd.DataFrame(df_concat.loc[df_concat['MI']== True].groupby(['YEAR','HOSP_DIVISION','AMONTH']).sum())
    MI_Regional = MI_Regional[['DISCWT']]
    MI_Regional = MI_Regional.rename(columns = {'DISCWT': 'MI Incidence'})
    print(f'created a dataframe with only regional MI admissions incidence : {file}' )

    # MI Total
    MI_Total = pd.DataFrame(df_concat.loc[df_concat['MI']== True].groupby(['YEAR','AMONTH']).sum())
    MI_Total = MI_Total[['DISCWT']]
    MI_Total = MI_Total.rename(columns = {'DISCWT': 'MI Incidence'})
    print(f'created a dataframe with only total MI admissions incidence : {file}' )

    #Both Merged to two dataframes
    Both_Regional = pd.merge(Influenza_Regional, MI_Regional, how = 'inner', on= ['YEAR','HOSP_DIVISION','AMONTH'])
    Both_Regional_Main = pd.concat([Both_Regional.reset_index(), Both_Regional_Main.reset_index()], ignore_index=True)
    Both_Regional_Main.set_index(['YEAR', 'HOSP_DIVISION', 'AMONTH'])
    Both_Regional_Main = Both_Regional_Main[['YEAR', 'HOSP_DIVISION', 'AMONTH', 'Influenza Incidence', 'MI Incidence']]    
    print(f'Merged data from: {file} into the main DataFrame' )
    Both_Total = pd.merge(Influenza_Total, MI_Total, how = 'inner', on= ['YEAR','AMONTH'])
    Both_Total_Main = pd.concat([Both_Total_Main.reset_index(), Both_Total.reset_index()], ignore_index=True)
    Both_Total_Main.set_index(['YEAR', 'AMONTH'])
    Both_Total_Main = Both_Total_Main[['YEAR', 'AMONTH', 'Influenza Incidence', 'MI Incidence']]
```
